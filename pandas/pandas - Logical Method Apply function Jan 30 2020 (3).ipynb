```python
def ChiefJobTitle(title):
    if 'CHIEF'.lower() in title.lower():
        return title
    else:
        return None
```

```python
def ChiefJobTitle1(title):
    if 'CHIEF'.lower() in title.lower():
        return True
    else:
        return False
```
